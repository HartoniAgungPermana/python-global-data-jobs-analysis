## **Salary Analysis of Data Jobs in United States**

### **Purpose**
- Identifying the distribution of salary range of top 6 most demanded data jobs in United States.
- Identifying the median salary of top 6 most demanded skills for data analysts in United States.
- Identifying the median salary of top 6 highest paid skills for data analysts in United States.

### **Method**
- Import libraries and load the dataset
- Clean and filter the dataset
- Aggregate the salary distribution of top 6 most demanded data jobs
- Aggregate the median salary of top 6 most demanded skills for data analyst jobs
- Aggregate the median salary of top 6 highest paid skills for data analyst jobs
- Plot the result

### **1. Import Libraries and Load The Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns
```

```python
df = pd.read_csv(rC:\Users\harto\OneDrive\Dokumen\DATA ANALYST PORTOFOLIO PROJECT\5. Data Jobs Analyis Using Python\dataset\data_jobs.csv')
df
```

### **2. Clean The Dataset**

```python
# Change data type into datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Change data type into list
def str_to_list (column) :
    if pd.notna(column) :
        return ast.literal_eval(column)
    
df['job_skills'] = df['job_skills'].apply(str_to_list)
```

```python
# Filter location only for United States
df_US = df[df['job_country'] == 'United States'] 
```

```python
# Get the top 6 job
top_6_job = df_US['job_title_short'].value_counts().sort_values(ascending = False).head(6).index.to_list()
top_6_job
```

### **3. Aggregate The Salary Distribution of Top 6 Most Demanded Data Jobs**

```python
# Filter by top 6 data jobs
df_boxplot = df_US[df_US['job_title_short'].isin(top_6_job)] 
df_boxplot
```

### **4. Plot The Boxplot of Salary Distribution**

```python
sns.set_theme(style = 'white')

sns.boxplot(data = df_boxplot, x = 'salary_year_avg', y = 'job_title_short')
sns.despine()

plt.title('Salary Distributions of Data Jobs Roles In United States')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 650000)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : f'${int(x/1000)}K'))
plt.show()
```

### **5. Aggregate Median Salary of Top 6 Most Demanded Skills and Highest Paid for Data Analyst Jobs**

```python
# Filter data for only data analyst
df_US_DA = df_US[df_US['job_title_short'] == 'Data Analyst'] 

# Explode the job skills
df_US_DA = df_US_DA.explode('job_skills')
df_US_DA
```

```python
# Most demanded skills
top_demand_skills = df_US_DA['job_skills'].value_counts().sort_values(ascending = False).head(7).index

# Most paid skills
top_paid_skills = df_US_DA.groupby('job_skills')['salary_year_avg'].median().sort_values(ascending = False).head(7).index
```

```python
df_barplot = df_US_DA.groupby('job_skills')['salary_year_avg'].median()
```

```python
# Aggregate median salary for most demand skills
df_most_demand = df_barplot.loc[top_demand_skills].reset_index(name = 'median_salary').sort_values(by ='median_salary' ,ascending = False)

# Aggregate median salary for most paid skills
df_top_paid = df_barplot.loc[top_paid_skills].reset_index(name = 'median_salary').sort_values(by ='median_salary' ,ascending = False)
```

### **6. Plot The Result**

```python
fig, ax = plt.subplots(2,1)

sns.set_theme(style = 'ticks')

# Plot most demand skills
sns.barplot(data = df_most_demand, x = 'median_salary', y = 'job_skills', ax = ax[0], palette = 'dark:b_r', hue ='median_salary')
ax[0].set_ylabel('')
ax[0].set_xlabel('Median Salary (USD)')
ax[0].set_title('Median Salary of Most Demanded Skills In Data Analyst Job')
ax[0].legend().remove()
ax[0].set_xlim(0, 200000)
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : f'${int(x/1000)}K'))

# Plot most paid skills
sns.barplot(data = df_top_paid, x = 'median_salary', y = 'job_skills', ax = ax[1], palette = 'dark:b_r', hue = 'median_salary')
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')
ax[1].set_title('Median Salary of Highest Paid Skills In Data Analyst Job')
ax[1].legend().remove()
ax[1].set_xlim(0, 200000)
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : f'${int(x/1000)}K'))

fig.tight_layout()

plt.show()
```
