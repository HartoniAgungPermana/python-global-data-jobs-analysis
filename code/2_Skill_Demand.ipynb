## **The most demanded skills on top 3 most popular data enthusiast jobs in United States**

### **Purpose**
- Identifying total number counts of most demanded skills in top 3 data enthusiast jobs in United States
- Identifying the likelihood percentage of most demanded skills in top 3 data enthusiast jobs in United States

### **Method**
- Import libraries and load the dataset
- Clean the dataset
- Aggregate the total counts of skills
- Aggregate the percentage of the likelihood of skills
- Plot the result

### **1. Import Libraries and Load The Dataset**

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Change the dataset path using yours if you want to rerun
df = pd.read_csv(r'C:\Users\harto\OneDrive\Dokumen\DATA ANALYST PORTOFOLIO PROJECT\5. Data Jobs Analyis Using Python\dataset\data_jobs.csv')
df
```

### **2. Clean and Filter The Dataset**

```python
# Change the data type into datetime
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) 
```

```python
# Change the datatype into list
def str_to_list(column):
    if pd.notna(column) :
        return ast.literal_eval(column)
    
df['job_skills'] = df['job_skills'].apply(str_to_list)
```

```python
# Get the top 3 most popular data jobs in United States
top_3_jobs = df[df['job_country'] == 'United States']['job_title_short'].value_counts().sort_values(ascending = False).head(3).index.to_list()
top_3_jobs
```

```python
# Filter the data for United States and top 3 jobs only
df_filtered = df[(df['job_country'] == 'United States') & (df['job_title_short'].isin(top_3_jobs))]
df_filtered
```

### **3. Aggregate The Total Counts of Skills**

```python
# Aggregate the total count of each jobs
df_total_job = df_filtered['job_title_short'].value_counts().reset_index(name = 'jobs_total')
df_total_job
```

```python
# Explode job skills
df_filtered = df_filtered[['job_title_short', 'job_skills']].explode('job_skills') 
df_filtered
```

```python
# Aggregate the total count of job skills for each data jobs
df_skill_count = df_filtered.groupby(['job_skills', 'job_title_short']).size() 
df_skill_count = df_skill_count.reset_index(name = 'skill_count')
df_skill_count = df_skill_count.sort_values(by = 'skill_count', ascending = False)

df_skill_count
```

### **4. Aggregate The Likelihood Percentage of Job Skills**

```python
# Aggregate the percentage of the likelihood of skills
df_final = pd.merge(df_skill_count, df_total_job, on = 'job_title_short', how = 'left')
df_final['percentage'] = (df_final['skill_count'] / df_final['jobs_total']) * 100
df_final
```

### **5. Plot The Total Counts of Most Demanded Skill**

```python
fig, ax = plt.subplots(3,1)

sns.set_theme(style = 'ticks')

for i, job_title in enumerate(top_3_jobs):
    df_plot = df_final[df_final['job_title_short'] == job_title].head(5)
    sns.barplot(data = df_plot, x = 'skill_count', y = 'job_skills', ax = ax[i], hue = 'skill_count', palette = 'dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0,50000)
    if i != 2:
        ax[i].set_xticks([])
    for n, v in enumerate(df_plot['skill_count']):
        ax[i].text(v + 1, n, v, va='center')

fig.suptitle('Counts of Skills Requested In United States Job Posting')
fig.tight_layout()
plt.show()
```

### **6. Plot The Likelihood Percentage of Most Demanded Skills**

```python
fig, ax = plt.subplots(3,1)

sns.set_theme(style = 'ticks')

for i, job_title in enumerate(top_3_jobs):
    df_plot = df_final[df_final['job_title_short'] == job_title].head(5)
    sns.barplot(data = df_plot, x = 'percentage', y = 'job_skills', ax = ax[i], hue = 'percentage', palette = 'dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0,80)
    ax[i].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos : f'{int(x)}%'))
    if i != 2:
        ax[i].set_xticks([])
    for n, v in enumerate(df_plot['percentage']):
        ax[i].text(v + 1, n, f'{v:.0f}%', va='center')

fig.suptitle('Percentage of Skills Requested In United States Job Posting')
fig.tight_layout()
plt.show()
```
